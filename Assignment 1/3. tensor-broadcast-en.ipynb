# Tensor vs Tensor: Same Shape
* Same shape: Perform operations on **corresponding** element pairs; also known as **element-wise operation**

````python
%%writefile dsatutorials/src/main.cpp

#include <iostream>
#include <iomanip>
#include <sstream>
#include <string>
using namespace std;

#include "sformat/fmt_lib.h"
#include "tensor/xtensor_lib.h"


void demo_operator(){
  xt::xarray<double> x = {
    {1, 3, 5, 7},
    {2, 4, 6, 8}
  };
  xt::xarray<double> y = {
    {10, 20, 30, 40},
    {60, 70, 80, 90}
  };


  cout << "x: " << endl;
  cout << x << endl;
  cout << "y: " << endl;
  cout << y << endl << endl;

  cout << "x + y: " << endl;
  cout << x + y << endl << endl;
  cout << "x - y: " << endl;
  cout << x - y << endl << endl;
  cout << "x * y: " << endl;
  cout << x * y << endl << endl;
  cout << "x / y: " << endl;
  cout << x / y << endl << endl;

}

int main(int argc, char** argv) {
    demo_operator();

    return 0;
}
````

# Tensor vs Tensor: Different Shapes
* When tensors have different shapes, they must be **broadcasted**. If **broadcasting** is successful, the operation can proceed; if not, a runtime error occurs.
* **Broadcasting** rules:
  * Step 1: Align the shapes to the right.
  * Step 2: Compare dimensions from right to left.
    * Step 2.1: If the sizes of the two dimensions are **equal**, proceed to the next dimension on the left.
      * Example:
        * Shape ```(6, 4)``` and ``` (5, 4)```: The rightmost dimensions are equal (both are 4) => proceed to the next dimension on the left, which is 6 and 5.
        * Shape ```(9, 6, 4)``` and ``` (6, 4)```: The second dimension from the right is equal (both are 6) => proceed to the leftmost dimension, which is 9 and **empty**.
    * Step 2.2: If the two dimensions are different and neither of the sizes is **1**, then an error occurs: **broadcasting cannot succeed**.
      * Example:
        * Shape: ```(5, 3, 4)``` and ``` (3, 5)```. The first dimension from the right is 4 and 5, which are different => broadcasting fails.
        * Shape: ```(5, 3, 4)``` and ``` (6, 4)```. The rightmost dimension matches (see Step 2.1). The next dimension is 3 vs 6, which causes failure.
    * Step 2.3: If the two dimensions are different, and one of the dimensions is **1**, then **broadcasting succeeds**. This means the block of data with dimension 1 is repeated as many times as the size of the other dimension.
      * Example:
        * Shape: ```(5, 3, 4)``` and ``` (5, 1, 4)```:
          * The rightmost dimension matches (4) => successful. Proceed to the next dimension.
          * The next dimension is 3 and 1: success; **broadcasting** means the data with shape ``` (5, 1, 4)``` is repeated 3 times in the second dimension from the right.
    * Step 2.4: If a dimension is **empty** (non-existent), **broadcasting succeeds**.
      * Example 1:
        * Shape: ```(5, 3, 4)``` and ``` (3, 4)```:
          * The rightmost dimension matches (4) => success. Proceed to the next dimension.
          * The next dimension is 3 and 3: success. Proceed to the next dimension.
          * The next dimension is 5 and **empty**. **Broadcasting** succeeds. This means the block of data ``` (3, 4)``` is repeated 5 times to form the shape ```(5, 3, 4)```.
      * Example 2:
        * Shape: ```(9, 5, 3, 4)``` and ``` (3, 4)```:
          * The rightmost dimension matches (4) => success. Proceed to the next dimension.
          * The next dimension is 3 and 3: success. Proceed to the next dimension.
          * The next dimension is 5 and **empty**. **Broadcasting** succeeds. This means the block of data ``` (3, 4)``` is repeated 5 times to form the shape ```(5, 3, 4)```.
          * The next dimension is 9 and **empty**. **Broadcasting** succeeds. This means the block of data ``` (5, 3, 4)``` (already broadcasted in the previous step) is repeated 9 times to form the shape ```(9, 5, 3, 4)```.

## Unsuccessful Case

````python
%%writefile dsatutorials/src/main.cpp

#include <iostream>
#include <iomanip>
#include <sstream>
#include <string>
using namespace std;

#include "sformat/fmt_lib.h"
#include "tensor/xtensor_lib.h"


void demo_broadcast(){
  xt::xarray<int> X = xt::arange<int>(4*3).reshape({4, 3});
  xt::xarray<int> y = xt::arange<int>(4);

  cout << "X: " << endl;
  cout << X << endl;
  cout << "y: " << endl;
  cout << y << endl << endl;

  try{
    cout << "X.shape(): " << shape2str(X.shape()) << endl;
    cout << "y.shape(): " << shape2str(y.shape()) << endl;
    cout << "X + y: " << endl;
    cout << X + y << endl << endl;
  }catch(const xt::broadcast_error& e){
    cout << "X + y: broadcast is unsucessful" << endl;
    cout << e.what() << endl << endl;
  }

}

int main(int argc, char** argv) {
    demo_broadcast();

    return 0;
}
````
